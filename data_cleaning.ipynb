```python
# drop NaN values
df = df.dropna()

# drop duplicates
df.drop_duplicates(inplace=True)

```

##### TASK: Filter out all events from Germany, with 50 km distance, from year 2022

```python
df_filtered = df[
    (df["Event distance/length"].isin(["50km"])) &
    (df["Year of event"] == 2022) &
    (df["Event name"].str.contains(r'\(GER\)'))].copy()
```

```python
df_filtered.head()
```

##### Data Cleaning 1 

```python
# remove 'h' 
df_filtered["Athlete performance"] = df_filtered["Athlete performance"].str.replace('h', '', regex=True)
```

```python
# remove "M" or "W" 
df_filtered["Athlete age category"] = df_filtered["Athlete age category"].str.replace('^[MW]', '', regex=True)
```

```python
df_filtered
```

##### Data Cleaning 2 

```python
# add column 'Athlete age'
df_filtered['Athlete age'] = 2022 - df_filtered['Athlete year of birth']

# fix types
df_filtered['Athlete age'] = df_filtered['Athlete age'].astype('Int64')
```

```python
# drop columns 
df_filtered.drop(columns = ['Athlete club', 'Athlete year of birth', 'Athlete age category', 'Year of event'])
```

##### Data Preparation

```python
# rename columns
df_filtered = df_filtered.rename(columns = {'Event dates': 'date',
                          'Event name':'run_name',
                          'Event distance/length':'run_length',
                          'Event number of finishers':'run_finishers',
                          'Athlete performance':'run_time',
                          'Athlete country':'athlete_country',
                          'Athlete gender':'athlete_gender',
                          'Athlete average speed':'athlete_avg_speed',
                          'Athlete ID':'athlete_id',
                          'Athlete age':'athlete_age'
                         })

# reorder columns

df_filtered = df_filtered[['date', 'run_name', 'run_length', 'run_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_country', 'run_time', 'athlete_avg_speed']]
```
